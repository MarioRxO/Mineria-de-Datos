```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('datos_estudiantes_decision.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data = df, x="horas_estudio", y="calificacion")
plt.title("Horas de estudio vs Calificacion")
plt.ylabel("Calificacion")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data = df, x="horas_sueno" , y="nivel_estres")
plt.title("Horas de sueño vs Nivel de estres")
plt.ylabel("Calificacion")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data = df, x="uso_redes" , y="calificacion")
plt.title("Uso de Redes vs Calificacion")
plt.ylabel("Calificacion")
plt.show()
```

```python
#perarson para las 3
person_estudio_cal = df['horas_estudio'].corr(df['calificacion'], method='pearson')
person_sueno_estres = df['horas_sueno'].corr(df['nivel_estres'], method='pearson')
person_redes_cal = df['uso_redes'].corr(df['calificacion'], method='pearson')

print(f"Estudio/Calificación: {person_estudio_cal}")
print(f"Sueño/Estrés: {person_sueno_estres}")
print(f"Redes/Calificación: {person_redes_cal}")

# .90 es casi perfecto
```
