```python
import pandas as pd
df = pd.read_csv('employees.csv')

```

Obtén todas las filas del depto "Eng" que además tengan perf_score >= 0.85, pero muestra solo las columnas: emp_id, name, city, perf_score, salary_mxn.

```python
df.loc[(df["dept"] == "Eng") & (df["perf_score"] >= 0.85),["emp_id", "name", "city", "perf_score", "salary_mxn"]]
```

Del DataFrame completo, obtén las filas 3 a 8 (incluyendo 3 y excluyendo 8), y solo las columnas 0, 1, 5 y 6 (por posición).

```python
df.iloc[3:8,[0,1,5,6]]
```

Calcula:Filtra empleados que estén en city CDMX o GDL, con warnings == 0 y absences <= 1. Muestra emp_id, name, city, warnings, absences.

```python
warnings = 0
absences = 1
ciudades = ["CDMX", "GDL"]
query = df.query("warnings == @warnings and city in @ciudades and absences >= @absences")
query[["emp_id", "name", "city", "warnings", "absences"]]
```

Calcula:
salario promedio (salary_mxn)
perf promedio (perf_score)
total de warnings (warnings)
en una sola salida (por ejemplo usando agg).

```python

df.agg({
    "salary_mxn": "mean",
    "perf_score": "mean",
    "warnings": "sum"
})
```

Por cada dept, calcula:
número de empleados
salario promedio
perf promedio
Ordena el resultado por salario promedio descendente.

```python
res = (
    df.groupby("dept")
      .agg(
          empleados=("dept", "count"),
          salario_promedio=("salary_mxn", "mean"),
          perf_promedio=("perf_score", "mean")
      )
      .sort_values("salario_promedio", ascending=False)
)
```
