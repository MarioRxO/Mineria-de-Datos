```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("streaming_platform.csv")
df
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df , x="hours_watched_per_week", y="age")
plt.title("Horas a la semana vs Edad")
plt.xlabel("Horas a la semana")
plt.ylabel("Edad")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df , x="account_creation_year", y="number_of_sessions")
plt.title("Año de creacion de la cuenta vs Año de creacion de la cuenta")
plt.xlabel("Año de creacion de la cuenta")
plt.ylabel("Número de sesiones")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df , x="days_since_last_login", y="account_creation_year")
plt.title("Último login vs Año de creación de la cuenta")
plt.xlabel("Último login")
plt.ylabel("Año de creación de la cuenta")
plt.show()
```

```python
pearson_1 = df['hours_watched_per_week'].corr(df['age'],method='pearson')
pearson_2 = df['account_creation_year'].corr(df['number_of_sessions'],method='pearson')
pearson_3 = df['days_since_last_login'].corr(df['account_creation_year'],method='pearson')

print("Pearson: \n")
print("Horas a la semana vs Edad: ",pearson_1)
print("Año de creacion de la cuenta vs Año de creacion de la cuenta: ",pearson_2)
print("Último login vs Año de creación de la cuenta: ",pearson_3)
```

```python
spearman_1 = df['hours_watched_per_week'].corr(df['age'],method='spearman')
spearman_2 = df['account_creation_year'].corr(df['number_of_sessions'],method='spearman')
spearman_3 = df['days_since_last_login'].corr(df['account_creation_year'],method='spearman')

print("Pearson: \n")
print("Horas a la semana vs Edad: ",spearman_1)
print("Año de creacion de la cuenta vs Año de creacion de la cuenta: ",spearman_2)
print("Último login vs Año de creación de la cuenta: ",spearman_3)
```
